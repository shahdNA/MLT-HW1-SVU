```python
# Import the library of Pandas.
import pandas as pd
```

```python
# Read The path of Excel file and save it in a variable called (df) --> it means DataFrame.
df = pd.read_excel(r"D:\MLT\data_clean.xlsx", engine="openpyxl")
```

```python
# Show columns
df.columns
```

```python
# Make path for every customer and save it in a variable.
path_for_every_customer = ( df.groupby(["account_id", "Country", "solution"])["types"].apply(list).reset_index() )
```

```python
# Show the first five rows.
path_for_every_customer.head()
```

```python
# Show paths for all.
path_for_every_customer
```

```python
# Transfor the types columns from List into string and include (-->) inside every cell.
path_for_every_customer["types"] = path_for_every_customer["types"].str.join(" --> ")
```

```python
# Show paths for every customer after modified.
path_for_every_customer
```

```python
# Show the first five rows.
path_for_every_customer.head()
```

```python
# Counts the number of iterations of each path and made new columns for iterations calling (duplicates).
path_duplicates = ( path_for_every_customer.groupby(["Country", "solution", "types"]).size().reset_index(name = "duplicates") ) 
```

```python
# Show the number of iterations of each path.
path_duplicates
```

```python
path_duplicates.head()
```

```python
# Make the top five paths.
top_five_paths = ( 
    path_duplicates.sort_values( ["Country", "solution", "duplicates"], ascending = [True, True, False] ).groupby(["Country", "solution"]).head(5) 
)
```

```python
# Show the top five paths.
top_five_paths
```

```python
# Transform (top fie paths) into Excel file.
top_five_paths.to_excel("top_five_paths.xlsx", index= False)
```
